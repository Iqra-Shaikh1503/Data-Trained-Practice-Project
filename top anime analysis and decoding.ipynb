```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

# 1.Scores ditribution

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Score'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

# 2.Popularity distribution

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Popularity'], bins=30, kde=True, color='red')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

# 3. Number of Members

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Members'], bins=30, kde=True, color='green')
plt.title('Distribution of Members')
plt.xlabel('Members')
plt.ylabel('Frequency')
plt.show()
```

# 4. Genres Distribution 

```python
top_genre = df['Genres'].str.split(',').explode().value_counts().head(15)
top_genre
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x=top_genre.values, y=top_genre.index, palette=('muted'))
plt.title('Top 15 genres Distibution')
plt.xlabel('COUNT')
plt.ylabel('Genres')
plt.show()
```

# 5. Score VS Popularity Scatter Plot

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Score', y='Popularity', data=df, alpha=0.7)
plt.title('Score VS Popularity')
plt.xlabel('Score')
plt.ylabel('Popularity')
plt.show()
```

# 6. Top 10 Producers by Count

```python
top_producers = df['Producers'].str.split(',').explode().value_counts().head(10)
top_producers
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x=top_producers.values, y=top_producers.index, palette='Blues_r')
plt.title('Top 10 Producers By count')
plt.xlabel('Count')
plt.ylabel('Producer')
plt.show()
```

# 7. Distribution of top anime show from each studio

```python
from collections import Counter
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing anime data
value_counts = df['Studios'].value_counts().head(20)
df_studios_count = pd.DataFrame({'Studios': value_counts.index, 'Animes_in_list': value_counts.values})

# Plotting the pie chart using Matplotlib
plt.figure(figsize=(12,10))
plt.pie(df_studios_count['Animes_in_list'], labels=df_studios_count['Studios'], autopct='%1.1f%%')
plt.title("Number of top anime from each studio")
plt.show()
```

# 8. Distribution of types of anime show

```python
value_counts = df['Type'].value_counts().head(6)
df_studios_count = pd.DataFrame({'Type': value_counts.index, 'Count': value_counts.values})

# Plotting the pie chart using Matplotlib
plt.figure(figsize=(12,10))
plt.pie(df_studios_count['Count'], labels=df_studios_count['Type'], autopct='%1.1f%%')
plt.title("Distribution of type of anime show")
plt.show()
```
